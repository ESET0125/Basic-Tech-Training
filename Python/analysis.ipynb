# Matplotlib

```python
pip install matplotlib
```

```python
import matplotlib .pyplot as plt # Import the matplotlib.pyplot module, commonly aliased as 'plt' for convenience
```

```python

x = [0,2,4,6,8] # Define a list 'x' representing the data for the x-axis.
y = [0,4,16,36,64] # Define a list 'y' representing the data for the y-axis.

fig,ax = plt.subplots() # Create a figure and a set of subplots. 'fig' represents the entire figure, and 'ax' represents the axes (the plotting area).
ax.plot(x, y, marker='o', label="Data Points") # Plot the data. 'x' and 'y' are the coordinates. 'marker='o'' adds circular markers to each data point. 'label="Data Points"' provides a label for the legend.
ax.set_title("Basic Components of Matplotlib Figure") # Set the title of the plot.
ax.set_xlabel("X-Axis") # Set the label for the x-axis.
ax.set_ylabel("Y-Axis") # Set the label for the y-axis.
ax.legend() # Display the legend, which uses the 'label' provided in the ax.plot() function.
plt.show() # Display the generated plot.
```

# combined example numpy, pandas and matpltlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Create a Numpy array with the data for the bar chart
# While a standard Python list can function like an array, NumPy arrays are designed for greater efficiency, performance, and functionality, especially when handling large datasets
arr = np.array([10,20,30,40,50,60,40,45,43,20,30])
arr2 = np.array([2,4,8,4,6,8,9,0,3,2,5])

# Convert the NumPy array into a Pandas DataFrame
# The dictionary keys become the column names, and the array becomes the column data.
df = pd.DataFrame({"Numbers": arr, "x": arr2})

# Generate a bar plot from the DataFrame
# `df.index` provides the position for each bar on the x-axis.
# `df["Numbers"]` provides the height of each bar on the y-axis.
plt.bar(df.x, df["Numbers"])

# Display the final plot
plt.show()
```
